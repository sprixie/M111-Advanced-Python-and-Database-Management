## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../data/company_sales_data.csv")
profit = df['total_profit']
months  = df['month_number']
plt.plot(months, profit)
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.xticks(months)
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

## Exercise 3
**Read all product sales data and show it  using a multiline plot**
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Expected output**

<img src="../data/sales_data.png" alt="sales_data" style="width: 500px;"/>


```python
facecream   = df ['facecream']
facewash   = df ['facewash']
toothpaste = df ['toothpaste']
bathingsoap   = df ['bathingsoap']
shampoo  = df ['shampoo']
moisturizer= df ['moisturizer']

plt.plot(months, facecream,label = 'Face cream Sales Data', marker='o', linewidth=3)
plt.plot(months, facewash, label = 'Face Wash Sales Data',  marker='o', linewidth=3)
plt.plot(months, toothpaste, label = 'Tooth paste Sales Data', marker='o', linewidth=3)
plt.plot(months, bathingsoap, label = 'Bathing soap Sales Data', marker='o', linewidth=3)
plt.plot(months, shampoo, label = 'Shampoo Sales Data', marker='o', linewidth=3)
plt.plot(months, moisturizer, label = 'Moisturizer Sales Data', marker='o', linewidth=3)

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(months)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
labels = ['Face cream', 'Fase wash', 'Tooth paste', 'Bathing soap', 'shampoo', 'Moisturizer']
totalsales   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(totalsales, labels=labels,  autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

**Expected output**

<img src="../data/Sales_data_of_a_facewash.png" alt="Sales_data_of_a_facewash" style="width: 700px;"/>


```python
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(months, bathingsoap, color='k', marker='o', linewidth=3)
axarr[0].set_title('Sales data of a Bathing soap')
axarr[1].plot(months, facewash, color='r', marker='o', linewidth=3)
axarr[1].set_title('Sales data of a facewash')

plt.xticks(months)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

```python
plt.plot([],[],color='m', label='face Cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(months, facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer, colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
